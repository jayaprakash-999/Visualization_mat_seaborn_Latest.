# seaborn

### seaborn tips inbult data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# tips  inbult data

tips =sns.load_dataset('tips')
```

```python
tips
```

## Univariate → “one variable” data visualization

```python
#strip plot
# jitter=False ---> Distribution of data

sns.stripplot(y= 'tip', data=tips,jitter=False)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot

sns.swarmplot(x='day', y='total_bill', data=tips)
plt.ylabel('tip ($)')
plt.show()


# Find more values in particular range
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex', orient='h')
plt.xlabel('tip ($)')
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

### Box and Violin plot

```python
#plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips, hue='time')
plt.ylabel('tip ($)')
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')

plt.show()
```

```python
plt.subplot(2,1,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')


plt.subplot(2,1,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
##Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

## Bivariate → “two variables” data visualization

```python
#Joint plot

sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
# Density plot
# kde = Kernal density estimator
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')
plt.show()
```

```python
#Pair plot

sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

```python
# delete null values
# Among 340 values reduce to 290

data.loc[data['Survived']==1,"Age"].dropna()
```

```python
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived");

# Histogram , Frequency curve ---> distribution plot
```

```python
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived");
```

```python
sns.pairplot(data.dropna());

# Only visable Numerical data
```

```python
# Pclass vs Survive
# with out any subplot ---> Faceplot
# col = column  ,  row = rows
#x axis = age   , Y axis = Surived

grid = sns.FacetGrid(data, col='Survived', row='Pclass', aspect=1.5)
grid.map(plt.hist, 'Age', alpha=0.5, bins=20)
grid.add_legend()

# aspect=1.5: This controls the aspect ratio of the individual plots within the grid, making them wider.
# alpha=.5: (Light colour) This sets the transparency of the histogram bars to 50%.
# legand gives pclass=1 | Survived = 0
```

```python

sns.distplot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()

plt.title("QueensTown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("Southampton")
sns.distplot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
#Most of the Passengers embarked from Southampton

import warnings
warnings.filterwarnings("ignore")
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
sns.distplot(not_survived,ax=axesbi[1,3])

# figbi size 2 ,4
# axesbi 16,10
# xlim=[0,1]
# barh ---> horizental bar
# xlim=[0,1] ---> Servived data
"""" xlim=[0,1]: Sets the x-axis limits to be between 0 and 1, likely because survival rate is a proportion."""
```
